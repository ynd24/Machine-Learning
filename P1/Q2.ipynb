```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
```

```python
df
```

```python
sepal_len = list(df['sepal length (cm)'])
sepal_wid = list(df['sepal width (cm)'])
petal_len = list(df['petal length (cm)'])
petal_wid = list(df['petal width (cm)'])
target = list(df['target'])
```

```python
i_setosa_seplen = sepal_len[0:50]
i_versicolour_seplen = sepal_len[50:100]
i_virginica_seplen = sepal_len[100:150]

three_fl_seplen = [i_setosa_seplen, i_versicolour_seplen, i_virginica_seplen]
plt.boxplot(three_fl_seplen, labels=['Iris Setosa sep_len', 'Iris Versicolour sep_len', 'Iris Virginica sep_len'])
plt.show
```

```python
setosa_sep_wid = sepal_wid[0:50]
versicolour_sep_wid = sepal_wid[50:100]
virginica_sep_wid = sepal_wid[100:150]

sepal_wid_three_fl = [setosa_sep_wid, versicolour_sep_wid, virginica_sep_wid]

plt.boxplot(sepal_wid_three_fl, labels=['Iris Setosa sep_wd', 'Iris Versicolour sep_wid', 'Iris Virginica sep_wid'])
plt.show()
```

```python
setosa_pet_len = petal_len[0:50]
versicolour_pet_len = petal_len[50:100]
virginica_pet_len = petal_len[100:150]

petal_len_three_fl = [setosa_pet_len, versicolour_pet_len, virginica_pet_len]

plt.boxplot(petal_len_three_fl, labels=['Iris Setosa pet_len', 'Iris Versicolour pet_len', 'Iris Virginica pet_len'])
plt.show()
```

```python
setosa_pet_wid = petal_wid[0:50]
versicolour_pet_wid = petal_wid[50:100]
virginica_pet_wid = petal_wid[100:150]
petal_wid_three_fl = [setosa_pet_wid, versicolour_pet_wid, virginica_pet_wid]
plt.boxplot(petal_wid_three_fl, labels=['Iris Setosa pet_wid', 'Iris Versicolour pet_wid', 'Iris Virginica pet_wid'])
plt.show()
```
